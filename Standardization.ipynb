```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Social_Network_Ads.csv')
df
```

```python
df=df.iloc[:, 2:]
```

```python
df.sample(5)
```

```python
# Train-test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased', axis=1), df['Purchased'], train_size=0.70, random_state=42)
```

```python
#Standardization using StandardScaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler() #creating object for standardScaler

#transforming train and test set
X_train_scaled= scaler.fit_transform(X_train)
X_test_scaled= scaler.transform(X_test)
```

```python
#Effect Of Scaling


# Create a figure and two subplots
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# Scatter plot on the first subplot
ax1.scatter(X_train['Age'], X_train['EstimatedSalary'])
ax1.set_title("Before Scaling")

# Scatter plot on the second subplot
ax2.scatter(X_train_scaled['Age'], X_train_scaled['EstimatedSalary'])
ax2.set_title("After Scaling")

# Show the plots
plt.show()

```
